# Extract


````python
import pandas as pd
from sqlalchemy import create_engine
````

````python
csv_file = "fulltimeSF 2.csv"
sf_jobs_df = pd.read_csv(csv_file)
sf_jobs_df.head()
````

````python
new_sf_jobs_df = sf_jobs_df[['position', 'company', 'location']].copy()
new_sf_jobs_df.head()
````

````python
csv_file2 = "Facebook_Salaries_2018.csv"
facebook_jobs_df = pd.read_csv(csv_file2)
facebook_jobs_df.head()
````

````python
new_facebook_jobs_df = facebook_jobs_df[["Job_Role", "Pay", "Location"]].copy()
new_facebook_jobs_df.head()
````

````python
csv_file3 = "Google_Salaries_2018.csv"
google_jobs_df = pd.read_csv(csv_file3)
google_jobs_df.head()
````

````python
new_google_jobs_df = facebook_jobs_df[["Job_Role", "Pay", "Location"]].copy()
new_google_jobs_df.head()
````

````python
csv_file4 = "SalesForce_Salaries_2018.csv"
salesforce_jobs_df = pd.read_csv(csv_file4)
salesforce_jobs_df.head()
````

````python
new_salesforce_jobs_df = facebook_jobs_df[["Job_Role", "Pay", "Location"]].copy()
new_salesforce_jobs_df.head()
````
